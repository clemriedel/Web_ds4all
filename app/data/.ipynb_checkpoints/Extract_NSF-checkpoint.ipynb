```python
import csv
import pandas as pd
```

```python
input_file = csv.DictReader(open("NSF_Bio_sample300.csv"))

text = [] 
number = []

for row in input_file:
    text.append(row["Abstract"])
    number.append(row["AwardedAmountToDate"])

number_cleaned=[]
for amount in number: 
    number_cleaned.append(float(
            str(amount).replace("$", "").replace(",", "").replace(".00", "")
                                )/1e6
                          )
df = pd.DataFrame({'$':number_cleaned, 'NSF Bio 300 samples':text})

df.to_csv('NSF_Bio_sample3002.csv')
```
